```python
import numpy as np
import matplotlib.pyplot as plt

# Time array
t = np.linspace(0, 2 * np.pi, 500)
omega = 1  # Angular frequency

# Complex exponential
z = np.exp(1j * omega * t)

# Real and imaginary parts
real_part = np.real(z)
imag_part = np.imag(z)

# Plotting
plt.figure(figsize=(12, 5))

# Real and Imaginary parts vs time
plt.subplot(1, 2, 1)
plt.plot(t, real_part, label='Re{e^{jωt}}', color='blue')
plt.plot(t, imag_part, label='Im{e^{jωt}}', color='orange')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Real and Imaginary Parts of $e^{j\\omega t}$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Circular plot in complex plane
plt.subplot(1, 2, 2)
plt.plot(real_part, imag_part, label='$e^{j\\omega t}$ trajectory')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('Re')
plt.ylabel('Im')
plt.gca().set_aspect('equal')
plt.title('Trajectory of $e^{j\\omega t}$ in the Complex Plane')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
print("specific to only e^jwt")
```

```python
# Compute the complex conjugate and the sum
z_conj = np.exp(-1j * omega * t)
z_sum = z + z_conj  # Should be purely real

# Extract real and imaginary parts of the sum
real_sum = np.real(z_sum)
imag_sum = np.imag(z_sum)

# Plotting the sum
plt.figure(figsize=(10, 4))

plt.plot(t, real_sum, label='Real part: Re{e^{jωt} + e^{-jωt}}', color='blue')
plt.plot(t, imag_sum, label='Imaginary part: Im{e^{jωt} + e^{-jωt}}', color='orange', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.title('Real and Imaginary Parts of $e^{j\\omega t} + e^{-j\\omega t}$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Compute the difference
z_diff = z - z_conj  # e^{jωt} - e^{-jωt}

# Real and imaginary parts
real_diff = np.real(z_diff)
imag_diff = np.imag(z_diff)

# Plotting the result
plt.figure(figsize=(10, 4))

plt.plot(t, real_diff, label='Real part: Re{e^{jωt} - e^{-jωt}}', color='blue')
plt.plot(t, imag_diff, label='Imaginary part: Im{e^{jωt} - e^{-jωt}}', color='orange', linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.title('Real and Imaginary Parts of $e^{j\\omega t} - e^{-j\\omega t}$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.show()
```

```python
def f_beta_score(precision, recall, beta):
    beta_squared = beta ** 2
    numerator = (1 + beta_squared) * (precision * recall)
    denominator = (beta_squared * precision) + recall
    return numerator / denominator

# Given values from the confusion matrix
precision = 0.75
recall = 0.80

# Compute F0.5 and F2
f_0_5 = f_beta_score(precision, recall, 0.5)
f_2 = f_beta_score(precision, recall, 2)

print(f"F0.5-score: {f_0_5:.4f}")
print(f"F2-score: {f_2:.4f}")
```
