```python
import numpy as np

def activation_function(v):
    return 1 if v > 0 else -1

def perceptron_learning_one_sample_per_iteration(
    X,
    y,
    learning_rate=5,
    max_iterations=10,
    start_index=1
):
    """
    Perceptron algorithm processing exactly one data sample per iteration,
    starting at 'start_index' in the dataset.
    Uses branching logic for weight updates (no explicit factor of y_actual).
    """
    num_samples, num_features = X.shape
    # Initialize weights to zero
    weights = np.zeros(num_features)

    print("Iteration | Weights         | x           | w^T*x   | Predicted | Actual | Action")
    print("-" * 80)

    for iteration in range(max_iterations):
        # Keep a copy of the old weights so we can print them first
        weight_prev = weights.copy()

        # Select index using (start_index + iteration) % num_samples
        i = (start_index + iteration) % num_samples
        x_i = X[i]
        y_actual = y[i]

        # Compute w^T x using OLD weights, then predict
        v = np.dot(weight_prev, x_i)
        y_pred = activation_function(v)

        # Decide how to update the weights
        action_str = "Correct"
        weight_new = weight_prev  # by default, no change

        if y_pred != y_actual:
            if y_pred == -1 and y_actual == 1:
                # Predicted -1, should be +1 --> add
                weight_new = weight_prev + 2 * learning_rate * x_i
                action_str = (
                    f"w({iteration+1}) = w({iteration}) + 2 * {learning_rate} * x({i})"
                )
            elif y_pred == 1 and y_actual == -1:
                # Predicted +1, should be -1 --> subtract
                weight_new = weight_prev - 2 * learning_rate * x_i
                action_str = (
                    f"w({iteration+1}) = w({iteration}) - 2 * {learning_rate} * x({i})"
                )

        # Print/log using the old weight (weight_prev)
        print(
            f"{iteration:9} | {weight_prev} | {x_i} | {v:7.2f} | "
            f"{y_pred:^9} | {y_actual:^6} | {action_str}"
        )

        # Now apply the new weights AFTER printing
        weights = weight_new

    return weights


# ------------------------------------------------------------------------------
# Example usage:
# ------------------------------------------------------------------------------
if __name__ == "__main__":
    # Example dataset (bias is part of x vector)
    X = np.array([
        [1, -1,  0],   # x(0)
        [1,  0, -2],   # x(1)
        [1,  1, -1],   # x(2)
        [1,  2,  1]    # x(3)
    ])

    # Class labels
    y = np.array([1, -1, -1, 1])

    # Run with a specific start index
    start_index = 0
    final_weights = perceptron_learning_one_sample_per_iteration(
        X, 
        y, 
        learning_rate=1,
        max_iterations=12, 
        start_index=start_index
    )

    print("\nFinal Weights:", final_weights)
```

```python
#1. Import Necessary Liobraries
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import OneHotEncoder
```

```python
#Import the dataset
iris_dataset = pd.read_csv("iris-data-clean.csv")

#Save the iris dataset as a dataframe
iris_df = pd.DataFrame(iris_dataset)
```

```python
# a.) modify the cleam data such that such that you have class =+1: for “iris-setosa”, and class
     # = -1 for others. Pick two of the input variables for the rest of the questions
# - Use one hot encoder

# Suppose 'iris_df' is your original Iris DataFrame
iris_df_classed = iris_df.copy()

# 1. Collapse all classes into "Iris-setosa" or "Not-setosa"
iris_df_classed['setosa_or_not'] = iris_df_classed['class'].apply(
    lambda c: 'Iris-setosa' if c == 'Iris-setosa' else 'Not-setosa'
)

# 2. Encode the helper column
encoder = OneHotEncoder(sparse=False)  
encoded = encoder.fit_transform(iris_df_classed[['setosa_or_not']])
encoded_col_names = encoder.get_feature_names_out(['setosa_or_not'])
encoded_df = pd.DataFrame(encoded, columns=encoded_col_names)

# 3. Overwrite 'class' with 1 = setosa, 0 = not-setosa
iris_df_classed['class'] = encoded_df['setosa_or_not_Iris-setosa'].astype(int)

# 4. Replace 0 with -1, preserving 1 as 1
iris_df_classed['class'] = iris_df_classed['class'].apply(lambda val: 1 if val == 1 else -1)

# (Optional) drop the helper column
iris_df_classed.drop(columns=['setosa_or_not'], inplace=True)

iris_df["class"].value_counts()
iris_df_classed["class"].value_counts()
```

## Linear Separable Check

```python
import matplotlib.pyplot as plt

# Ensure the "class" column is numeric if it's an object type
#iris_df["class"] = iris_df["class"].astype(int)

# Separate data points based on class values
positive_class = iris_df_classed[iris_df_classed["class"] == 1]
negative_class = iris_df_classed[iris_df_classed["class"] == -1]

# Plot sepal_length_cm vs petal_length_cm with different markers for classes
plt.figure(figsize=(8, 6))
plt.scatter(
    positive_class["sepal_length_cm"], positive_class["petal_length_cm"],
    color="blue", marker="o", label="Class +1 (Iris-setosa)"
)
plt.scatter(
    negative_class["sepal_length_cm"], negative_class["petal_length_cm"],
    color="red", marker="x", label="Class -1 (Other)"
)

# Labels and title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Scatter Plot of Sepal Length vs Petal Length by Class")

# Add legend and grid for clarity
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()

print(" After analyzing the graph, the data points are linearly separable")
```

```python
#Scatter Plot of all combinations 
import itertools
import matplotlib.pyplot as plt

# Define feature names
features = ["sepal_length_cm", "sepal_width_cm", "petal_length_cm", "petal_width_cm"]

# Generate all possible 2-feature combinations (4 choose 2 = 6)
feature_combinations = list(itertools.combinations(features, 2))

# Ensure the class column is numeric
iris_df_classed["class"] = iris_df_classed["class"].astype(int)

# Set up subplots for 6 scatter plots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()  # Flatten the axes array for easy indexing

# Loop through each feature combination and plot it
for i, (x_feature, y_feature) in enumerate(feature_combinations):
    ax = axes[i]

    # Scatter plot for each class
    for class_label, marker, color in zip([1, -1], ['o', 'x'], ['blue', 'red']):
        subset = iris_df_classed[iris_df_classed["class"] == class_label]
        ax.scatter(subset[x_feature], subset[y_feature], marker=marker, color=color, label=f"Class {class_label}")

    # Labels and title
    ax.set_xlabel(x_feature)
    ax.set_ylabel(y_feature)
    ax.set_title(f"Scatter Plot: {x_feature} vs {y_feature}")
    ax.legend()
    ax.grid(True, linestyle="--", alpha=0.6)

# Adjust layout and show plots
plt.tight_layout()
plt.show()

print("All combinations are linearly separable")
```

```python
# Split the data into training and test 

from sklearn.model_selection import train_test_split
from perceptron import Perceptron

# Split into training and testing sets
iris_x_train, iris_x_test, iris_y_train, iris_y_test = train_test_split(
    iris_df_classed.drop(columns=["class"]),  # Features (X)
    iris_df_classed["class"],  # Labels (y)
    test_size=0.10,  # 25% test data
    random_state=np.random.randint(100)
)

# Convert to the correct numerical types (ensures no dtype errors)
iris_x_train = iris_x_train.astype(float).values
iris_x_test = iris_x_test.astype(float).values
iris_y_train = iris_y_train.astype(int).values
iris_y_test = iris_y_test.astype(int).values
```

```python
## Train the data on the perceptron 
# Initialize Perceptron with learning rate and number of iterations
pn = Perceptron(rate=.03, niter=10)

# Train the perceptron
pn.fit(iris_x_train, iris_y_train)
```

```python
# Check misclassification per epoch
print("Misclassification errors per epoch:", pn.errors)
```

```python
# Predict on test data
y_pred = pn.predict(iris_x_test)

# Calculate accuracy
accuracy = np.mean(y_pred == iris_y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python

# Plot training accuracy as a function of epochs
import matplotlib.pyplot as plt

# Compute training accuracy per epoch
total_samples = len(iris_y_train)
accuracy_per_epoch = [(total_samples - err) / total_samples for err in pn.errors]

plt.figure(figsize=(8, 5))
plt.plot(range(1, pn.niter + 1), accuracy_per_epoch, marker='o', linestyle='-')
plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.title("Training Accuracy vs. Number of Epochs")
plt.grid(True)
plt.show()
```
