```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x1 and x2
x1 = np.linspace(-5, 5, 10)
x2 = np.linspace(-5, 5, 10)
X1, X2 = np.meshgrid(x1, x2)

# Compute x3 from the decision boundary equation: 2x1 + x2 - x3 = 5
X3 = 2*X1 + X2 - 5

# Plot the plane
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, X3, color='cyan', alpha=0.6)

# Labels
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("x3")
ax.set_title("Perceptron Decision Boundary")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data points
positive_class = np.array([[0,0], [1,0], [1,-1], [1,1], [0,1]])
negative_class = np.array([[-1,0], [-1,1], [-1,-1]])

# Plot data points
plt.scatter(positive_class[:,0], positive_class[:,1], color='blue', label="Class +1")
plt.scatter(negative_class[:,0], negative_class[:,1], color='red', label="Class -1")

# Plot decision boundary (x1 = -0.5)
plt.axvline(x=-0.5, color='black', linestyle='--', label="Decision Boundary")

# Labels and legend
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Linearly Separable Dataset")
plt.legend()
plt.grid()
plt.show()
```
