```python
#Split the data into training and test data
X_train, X_test, y_train, y_test = train_test_split(X_, df_.Target2, shuffle=True, test_size=.25, random_state=42)
```

```python
#1. Logistic Regression No penalty (Multiple Logistic Regression)

#Logisitic Regression Classifier for Mutliclass
clf = LogisticRegression(penalty=None, multi_class='auto', solver='newton-cg')

# Fit the Model
clf.fit(X_train, y_train)

#------------------------------------------------------------------------------#
#Cross Validation Portion
#Assert the cross validation method: 
loo = LeaveOneOut()
loo.get_n_splits(X_train)
scores = list()
cv_scores_multReg = cross_val_score(clf, X_train, y_train, cv=loo, n_jobs = -1).mean() 
#------------------------------------------------------------------------------#

#Create the Predicitions based on the trained model
y_pred_multReg_test = clf.predict(X_test) # Test Performance
y_pred_multReg_train = clf.predict(X_train) # Train Performance
```

```python
#1. Logisitc Regression (Multiple Input) Results

print('----------------------------------------------------------------------------------') 
print(clf)
print('classes: ',clf.classes_)
print('coefficients: ',clf.coef_)
print('intercept :', clf.intercept_)
print('----------------------------------------------------------------------------------') 
print("Coef:",  len(clf.coef_)) 
print('----------------------------------------------------------------------------------') 
accuracy_multReg_test = accuracy_score(y_test, y_pred_multReg_test)
print("Accuracy for Log Reg Test:", accuracy_multReg_test)
print('----------------------------------------------------------------------------------') 
accuracy_multReg_train = accuracy_score(y_train, y_pred_multReg_train)
print("Accuracy for Log Reg Train:", accuracy_multReg_train)
print('----------------------------------------------------------------------------------') 
print('Cross Validation Score:', cv_scores_multReg)
print('----------------------------------------------------------------------------------') 
```

```python
# 1. Classification Reports for Logistic Regression
print('----------------------------------------------------------------------------------') 
print('Classification Report for Logistic Regression Test')
print('----------------------------------------------------------------------------------')
print(classification_report(y_test, y_pred_multReg_test, target_names=['dropout', 'graduate', 'enrolled']))

print('----------------------------------------------------------------------------------') 
print('Classification Report for Logistic Regression Train')
print('----------------------------------------------------------------------------------')
print(classification_report(y_train, y_pred_multReg_train, target_names=['dropout', 'graduate', 'enrolled']))
```
