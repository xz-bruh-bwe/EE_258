```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0, z)

# Define their derivatives
def sigmoid_derivative(z):
    sig = sigmoid(z)
    return sig * (1 - sig)

def tanh_derivative(z):
    return 1 - np.tanh(z)**2

def relu_derivative(z):
    return np.where(z > 0, 1, 0)

# Generate values for plotting
z = np.linspace(-5, 5, 400)

# Compute function values
sigmoid_vals = sigmoid(z)
tanh_vals = tanh(z)
relu_vals = relu(z)

# Compute derivative values
sigmoid_deriv_vals = sigmoid_derivative(z)
tanh_deriv_vals = tanh_derivative(z)
relu_deriv_vals = relu_derivative(z)

# Plot Sigmoid and its derivative
plt.figure(figsize=(5, 3))
plt.plot(z, sigmoid_vals, label="Sigmoid", color='blue')
plt.title("Sigmoid Activation Function")
plt.xlabel("z")
plt.ylabel("Activation")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(5, 3))
plt.plot(z, sigmoid_deriv_vals, label="Sigmoid Derivative", color='blue')
plt.title("Derivative of Sigmoid Function")
plt.xlabel("z")
plt.ylabel("Derivative")
plt.legend()
plt.grid()
plt.show()

# Plot Tanh and its derivative
plt.figure(figsize=(5, 3))
plt.plot(z, tanh_vals, label="Tanh", color='red')
plt.title("Tanh Activation Function")
plt.xlabel("z")
plt.ylabel("Activation")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(5, 3))
plt.plot(z, tanh_deriv_vals, label="Tanh Derivative", color='red')
plt.title("Derivative of Tanh Function")
plt.xlabel("z")
plt.ylabel("Derivative")
plt.legend()
plt.grid()
plt.show()

# Plot ReLU and its derivative
plt.figure(figsize=(5, 3))
plt.plot(z, relu_vals, label="ReLU", color='green')
plt.title("ReLU Activation Function")
plt.xlabel("z")
plt.ylabel("Activation")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(5, 3))
plt.plot(z, relu_deriv_vals, label="ReLU Derivative", color='green')
plt.title("Derivative of ReLU Function")
plt.xlabel("z")
plt.ylabel("Derivative")
plt.legend()
plt.grid()
plt.show()
```

```python
#Filter data 
# --- FILTERING EXAMPLES ---

# 1. Filter rows where housing_median_age is greater than 30
filtered_df1 = housingdf[housingdf['housing_median_age'] > 30]
print("\nFiltered DataFrame (housing_median_age > 30):\n", filtered_df1)

# 2. Filter rows where median_income is greater than 6 AND total_rooms > 1000
filtered_df2 = housingdf[(housingdf['median_income'] > 6) & (housingdf['total_rooms'] > 1000)]
print("\nFiltered DataFrame (median_income > 6 and total_rooms > 1000):\n", filtered_df2)

# 3. Filter rows where ocean_proximity is 'NEAR BAY'
filtered_df3 = housingdf[housingdf['ocean_proximity'] == 'NEAR BAY']
print("\nFiltered DataFrame (ocean_proximity = 'NEAR BAY'):\n", filtered_df3)

# 4. Filter rows where population is between 300 and 1000
filtered_df4 = housingdf[(housingdf['population'] >= 300) & (housingdf['population'] <= 1000)]
print("\nFiltered DataFrame (population between 300 and 1000):\n", filtered_df4)
```

```python
#reset Data

# Load the dataset (assuming housingdf is already created or loaded)
# housingdf = pd.read_csv('your_dataset.csv')  # Uncomment if loading from a CSV

# Apply filtering conditions based on your column names
filtered_df = housingdf.loc[
    (housingdf['longitude'] > -120) &  # Example condition for longitude
    (housingdf['latitude'] > 35) &  # Example condition for latitude
    (housingdf['median_income'] > 4)  # Example condition for median income
]

# Reset index
filtered_df.reset_index(drop=True, inplace=True)

# Display the filtered DataFrame
print(filtered_df)

# Save the filtered DataFrame to a CSV file
filtered_df.to_csv('filtered_data.csv', index=False)  # index=False prevents adding an extra index column
```

```python

# --- GROUPBY EXAMPLES ---

# 1. Group by 'ocean_proximity' and calculate the mean for all numeric columns
grouped_mean = housingdf.groupby('ocean_proximity').mean()
print("\nGrouped Mean Statistics:\n", grouped_mean)

# 2. Group by 'ocean_proximity' and calculate the sum for all numeric columns
grouped_sum = housingdf.groupby('ocean_proximity').sum()
print("\nGrouped Sum Statistics:\n", grouped_sum)

# 3. Group by 'ocean_proximity' and get multiple aggregations (mean, min, max)
grouped_agg = housingdf.groupby('ocean_proximity').agg({
    'median_income': ['mean', 'min', 'max'],
    'median_house_value': ['mean', 'min', 'max'],
    'total_rooms': ['sum', 'mean'],
    'total_bedrooms': ['sum', 'mean']
})
print("\nGrouped Aggregation (Mean, Min, Max, Sum):\n", grouped_agg)

# 4. Save the grouped statistics to a CSV file
grouped_agg.to_csv('grouped_statistics.csv')
```
