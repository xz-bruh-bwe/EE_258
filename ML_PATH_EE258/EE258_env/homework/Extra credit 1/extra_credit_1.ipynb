```python
# Define matrices
A = np.array([[1, 2, 4], [3, 4, 2]])
B_a = np.array([[4], [2]])  # (2x1)
B_b = np.array([[4], [5], [2]])  # (3x1)
B_c = np.array([4, 2])  # (1x2)
B_d = np.array([[6, 2, 4]]).T  # (3x1) Transposed to match dimensions
C_d = np.array([[5], [2], [1]])  # (3x1)

# Function to perform matrix multiplication and handle invalid cases
def matrix_multiply(A, B, label):
    try:
        result = np.dot(A, B)
        print(f"Result of {label}:\n{result}\n")
    except ValueError as e:
        print(f"Operation {label} is invalid: {e}\n")

# Perform matrix multiplications
matrix_multiply(A, B_a, "a")  # Invalid (2x3) * (2x1), mismatch
matrix_multiply(A, B_b, "b")  # Valid (2x3) * (3x1) -> (2x1)
matrix_multiply(A, B_c, "c")  # Invalid (2x3) * (1x2), mismatch
matrix_multiply(B_d.T, C_d, "d")  # Valid (1x3) * (3x1) -> (1x1)
```

## Problem 3

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of vectors
n = 1000  

# Function to generate random vectors with a given L2 norm
def generate_l2_norm_vectors(n, norm_value):
    vectors = np.random.randn(n, 2)  # Generate random vectors
    norms = np.linalg.norm(vectors, axis=1, keepdims=True)  # Compute L2 norms
    return (vectors / norms) * norm_value  # Scale to desired norm

# Function to generate random vectors with a given L1 norm
def generate_l1_norm_vectors(n, norm_value):
    vectors = np.random.randn(n, 2)  # Generate random vectors
    abs_sums = np.sum(np.abs(vectors), axis=1, keepdims=True)  # Compute L1 norms
    return (vectors / abs_sums) * norm_value  # Scale to desired norm

# Generate vectors for each case
vectors_a = generate_l2_norm_vectors(n, 1)  # ||x||_2 = 1
vectors_b = generate_l1_norm_vectors(n, 1)  # ||x||_1 = 1
vectors_c = generate_l2_norm_vectors(n, 2)  # ||x||_2 = 2
vectors_d = generate_l1_norm_vectors(n, 2)  # ||x||_1 = 2

# Plot vectors
plt.figure(figsize=(8, 8))
plt.scatter(vectors_a[:, 0], vectors_a[:, 1], color='red', alpha=0.5, label=r"$||x||_2 = 1$")
plt.scatter(vectors_b[:, 0], vectors_b[:, 1], color='blue', alpha=0.5, label=r"$||x||_1 = 1$")
plt.scatter(vectors_c[:, 0], vectors_c[:, 1], color='green', alpha=0.5, label=r"$||x||_2 = 2$")
plt.scatter(vectors_d[:, 0], vectors_d[:, 1], color='purple', alpha=0.5, label=r"$||x||_1 = 2$")

# Formatting the plot
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.legend()
plt.title("Random Vectors with Different Norm Constraints")
plt.xlabel("x-coordinate")
plt.ylabel("y-coordinate")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
n = 100  # Number of samples
sigma_squared = 1  # Variance of noise
a = 2  # Scaling factor
b = 3  # Intercept

# Step (a): Generate Gaussian random vector X with mean 0 and variance 1
X = np.random.normal(0, 1, (n, 1))

# Step (b): Create vector Y using linear mapping Y = aX + b
Y = a * X + b

# Step (c): Create a noisy version of Y using Y + Z, where Z is Gaussian noise with variance σ²
Z = np.random.normal(0, np.sqrt(sigma_squared), (n, 1))
Y_noisy = Y + Z

# Step (d): Compute correlation coefficient between X and Y_noisy
correlation = np.corrcoef(X.flatten(), Y_noisy.flatten())[0, 1]

# Step (e): Scatter plot of Y_noisy vs. X
plt.figure(figsize=(8, 6))
plt.scatter(X, Y_noisy, color='blue', alpha=0.5, label='Noisy Y vs. X')
plt.plot(X, Y, color='red', label='True Relationship (Y=aX+b)')
plt.xlabel('X')
plt.ylabel('Y (Noisy)')
plt.title('Scatter Plot of Y_noisy vs. X')
plt.legend()
plt.grid(True)
plt.show()

# Display correlation coefficient
correlation
```
